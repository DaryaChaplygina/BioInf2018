```python
import numpy as np
from scipy.stats import gaussian_kde
from matplotlib import pyplot as plt
%matplotlib inline
```

# 4M4

```python
def plot_estimations(alpha, beta):
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, 4), [alpha * x + beta for x in range(1, 4)], lw=0.3, c='black')
    plt.show()
    
    plt.figure(figsize=(10, 5))
    plt.title('Heights distribution for the 1st year')
    h1 = alpha + beta
    gkde = gaussian_kde(h1)
    plt.plot(range(250), [gkde(i) for i in range(250)])
```

```python
N = 100
alpha = np.random.normal(5, 2, N)
beta = np.random.normal(130, 15, N)
```

```python
plot_estimations(alpha, beta)
```

```python
alpha1 = np.random.chisquare(5, N)
beta1 = np.random.normal(130, 15, N)
plot_estimations(alpha1, beta1)
```
